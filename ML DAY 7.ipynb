# Label encoder

```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["python","ml","sql","aws","DL","power bi","python","ml","sql","aws","DL","power bi"]})
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(df["name"])
```

```python
df["en_name"] = le.fit_transform(df["name"])
```

```python
df
```

```python
new_data = df
new_data.to_csv("label_encoder.csv",index=False)
```
